```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('iris.csv')
```

```python
#scatter plot for Sepal Length and Width
colors=['red','orange','blue'] # array for colors
species = ['setosa', 'virginica', 'versicolor'] # array for labels
# put data in to scatter plot
for i in range(3):
    x = df[df['species'] == species[i]]
    plt.scatter(x['sepal_length'],x['sepal_width'], c=colors[i], label=species[i])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
```

```python
from sklearn.datasets import load_iris  # import built-in dataset
from sklearn.neighbors import KNeighborsClassifier # import KNNClassifer
import numpy 
from sklearn.model_selection import train_test_split # import Model Selection
irisDataset = load_iris() # import train and test data from iris dataset
Data_train, Data_test, Target_train, Target_test = train_test_split(irisDataset["data"], irisDataset["target"], random_state=0) # divide your data samples into train and test
data_to_predict = numpy.array([[5, 2.9, 1, 0.2]]) # a test sample
knn = KNeighborsClassifier(n_neighbors=7) # set a number for K _ Neighbors
knn.fit(Data_train, Target_train) # train your model
predicted_species = knn.predict(data_to_predict) # test your model with a random data to see the accuracy
print("The most probable species the sample will be is " + irisDataset["target_names"][predicted_species][0] 
	+ ". " + "The machine was tested and got {:.2f}% of its predictions right.".format(knn.score(Data_test, Target_test) * 100)) 
```
